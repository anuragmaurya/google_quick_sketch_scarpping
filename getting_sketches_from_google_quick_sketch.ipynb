```python
import sys
from selenium import webdriver
import time
import os

jsScript = """
        function move_up(element) {
            element.scrollTop = element.scrollTop -243;
        }

        function move_down(element) {
            element.scrollTop = element.scrollTop + 243;
        }

        move_down(arguments[0]);
        move_down(arguments[0]);
        """
def save_screenshots_of_google_image_category(image_category,no_of_screenshots):
    
    """
    This function goes in quickdraw of google and collects sketch of a particular category 
    as screenshot of the page.
    And saves theme in google_sketch directory/image_category
    The list of image categories are already given in google_sketch_cat.txt
    """
    
    dire = './google_sketch/'+image_category
    url = 'https://quickdraw.withgoogle.com/data/'+str(image_category)
    j=1
    if os.path.isdir(dire):
        return 0

    else:
        try:
            driver = webdriver.Chrome(executable_path=r"driver1.exe")  ##This is given driver path manually.
            driver.get(url)
            
            centerPanel = driver.find_element_by_id("scroller")
            
        except:
            return 0
        for i in range(no_of_screenshots):
            if j==0:
                #print('bhai 3')
                break
                

            try:
                try:
                    os.listdir(dire)
                    #print('bhai 4')
                except:
                    os.mkdir(dire)
                    #print('bhai 5')
                time.sleep(2.5)
                #print('bhai 6')
                file_name = dire +'/screenshot'+ str(i+1)+'.jpg'
                driver.save_screenshot(file_name) 
                print(str(i+1)+'th screenshot save for',image_category)
                driver.execute_script(jsScript, centerPanel)
            except:
                break
    driver.close()
    return 0
            
f = open('google_sketch_cat.txt','r')
cat_ls = []
for k in f:
    a= k[:-1]
    cat_ls.append(a.replace(' ','_'))
cat_ls.reverse()


for cat in cat_ls:
    save_screenshots_of_google_image_category(cat,250)
        
    
```
